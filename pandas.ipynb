# Pandas.DataFrame class structure
class pandas.DataFrame(data=None, index=None, columns=None, copy=None)  
Series -> 1 dimension, DataFrame -> 2 dimension array

```python
import numpy as np
import pandas as pd
```

```python
np.random.seed(0)
arr = np.random.randint(10, size = (2, 2)) # 2*2 random array
print(arr)
```

```python
df1 = pd.DataFrame(arr, copy = False)
df2 = pd.DataFrame(arr, copy = True)
```

```python
arr[0,0] = 99
print(df1)
```

```python
print(df2)
```

```python
data = {'A' : [1,2], 'B' : [3,4]}
df = pd.DataFrame(data = data)
print(df)
```

```python
data = np.array([[1,2],[3,4]])
df = pd.DataFrame(data = data, index = ['row1', 'row2'], columns = ['col1', 'col2'])
print(df)
```

### add

```python
data = [[1,10,100], [2,20,200], [3,30,300]]
col = ['col1', 'col2', 'col3']
row = ['row1', 'row2', 'row3']
df = pd.DataFrame(data = data, index = row, columns = col)
print(df)
```

```python
result = df.add(1)
print(result)
```

```python
result = df + 1
print(result)
```

```python
data2 = [[3], [4], [5]]
df2 = pd.DataFrame(data = data2, index = ['row1', 'row2', 'row3'], columns = ['col1'])
print(df2)
```

```python
result = df.add(df2)
print(result)
```

```python
result = df.add(df2, fill_value = 0)
print(result)
```

### sub

```python
data = [[1,10,100], [2,20,200], [3,30,300]]
col = ['col1', 'col2', 'col3']
row = ['row1', 'row2', 'row3']
df = pd.DataFrame(data = data, index = row, columns = col)
print(df)
```

```python
result = df.sub(1)
print(result)
```

```python
result = df - 1
print(result)
```

```python
data2= [[3], [4], [5]]
col = ['col1']
row = ['row1', 'row2', 'row3']
df2 = pd.DataFrame(data = data2, index = row, columns = col)
print(df2)
```

```python
result = df.sub(df2)
print(result)
```

```python
result = df.sub(df2, fill_value = 0)
print(result)
```
