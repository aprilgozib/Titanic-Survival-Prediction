# numpy - array, shape, dtype

```python
import numpy as np
```

```python
array1 = np.array([1,2,3])
print('array 1 = ', array1)
print('array 1 type = ', type(array1))
```

```python
print('array 1 shape = ', array1.shape)
```

```python
list = [[1,2,3], [4,5,6]]
array2 = np.array(list)
print('array 2 = ', array2)
print('array 2 shape = ', array2.shape)
```

```python
array3 = np.array([[1,2,3]])
print('array 3 = ', array3)
print('array 3 shape = ', array3.shape)
```

```python
print('array 1 : {0} dimension, array 2 : {1} dimension, array 3 : {2} dimension'.format(array1.ndim, array2.ndim, array3.ndim))
```

```python
list_type = [1,2,3.0]
array_type = np.array(list_type)
print(array_type, 'data type : ', array_type.dtype)
```

```python
array_type = array_type.astype('int32')
print('array_type data type : ', array_type.dtype)
```

# arange, zeros, ones

```python
print('ex 1 : ', np.arange(10))
```

```python
print('ex 2 : ', np.arange(3,10))
```

```python
print('ex 3 : ', np.arange(1, 10, 2))
```

```python
zero_array = np.zeros((3, 2), dtype = 'int32')
print(zero_array)
print('data shape : ', zero_array.shape)
print('data tyep : ', zero_array.dtype)
```

```python
one_array = np.ones((3, 2))
print(one_array)
print('data shape : ', one_array.shape)
print('data type : ', one_array.dtype)
```

```python
array = np.arange(10)
print('array : \n', array)
array1 = array.reshape(2,5)
print('array1 : \n', array1)
array2 = array.reshape(5,2)
print('array2 : \n', array2)
```

```python
array3 = array.reshape(-1, 2)
print('array3 shape : ', array3.shape)
array4 = array.reshape(2, -1)
print('array4 shape : ', array4.shape)
```

```python
array5 = array.reshape(-1, 4)
```

```python
array = np.arange(1, 10)
print('array : ', array)
print('3rd element in array : ', array[2])
print('2nd element data type : ', type(array[1]))
```

```python
print('array[1:4] : ', array[1:4])
print('data type : ', type(array[1:4]))
```

```python
print(array[:6])
print(array[2:])
print(array[:])
```

```python
array = array.reshape(3,3)
print('array : \n', array)
print('array[0:2, 0:2] : \n', array[0:2, 0:2])
print('array[1:3, :] : \n', array[1:3, :])
print('array[:2, 1:] : \n', array[:2, 1:])
```

```python
print('array : \n', array)
print('array[[0,2], :] : \n', array[[0,2], :])
print('array[[0,2]] : \n', array[[0,2]])
print('array[[0,2], 1:] : \n', array[[0,2], 1:])
```

```python
array = np.arange(1, 10)
array1 = array[array > 4]
array2 = array[(2 <= array) & (array < 7)]

print('array : ', array)
print('array1 : ', array1)
print('array2 : ', array2)
```

```python
original_array = np.array([3,1,9,5])
print('original : ', original_array)
```

```python
sort_array1 = np.sort(original_array)
print('np.sort() -> : ', sort_array1)
print('original : ', original_array)
```

```python
sort_array2 = original_array.sort()
print('original_array.sort() : ', sort_array2)
print('original : ', original_array)
```

```python
print('decrease : ', np.sort(original_array)[::-1])
```

```python
array = np.array([[8, 12], [7, 1]])
sort_array_axis0 = np.sort(array, axis = 0)
print('row sort : ', sort_array_axis0)

sort_array_axis1 = np.sort(array, axis = 1)
print('col sort : ', sort_array_axis1)
```

```python
original_array = np.array([3,1,9,5])
original_index = np.argsort(original_array)
print('index of sorted array : ', original_index)
print('type of index : ', type(original_index))
```

```python
print('array sort using argsort : ', original_array[original_index])
```

```python
name_array = np.array(['john', 'mike', 'sarah', 'kate', 'samuel'])
score_array = np.array([78, 95, 84, 98, 88])
score_index = np.argsort(score_array)[::-1]
print('score index : ', score_index)
print('high score order : ', name_array[score_index])
```

```python
A = np.array([[1,2,3], [4,5,6]])
B = np.array([[7,8],[9,10],[11,12]])
print('dot : \n', np.dot(A,B))
```

```python
A = np.array([[1,2], [3,4]])
print('A transpose : \n', np.transpose(A))
```
